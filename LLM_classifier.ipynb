```python
import os
import base64
import requests
import pandas as pd

def query_images_in_folder(folder_path, output_csv):
    API_KEY = os.getenv('AZURE_OPENAI_KEY')
    headers = {
        "Content-Type": "application/json",
        "api-key": API_KEY,
    }
    ENDPOINT = "https://dataextraction.openai.azure.com/openai/deployments/gpt-4o/chat/completions?api-version=2024-02-15-preview"

    # Initialize an empty DataFrame to store the results
    df = pd.DataFrame(columns=["filename", "response"])

    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
            image_path = os.path.join(folder_path, filename)
            image_filename = os.path.basename(image_path)
            encoded_image = base64.b64encode(open(image_path, 'rb').read()).decode('ascii')

            payload = {
                "messages": [
                    {
                        "role": "system",
                        "content": [
                            {
                                "type": "text",
                                "text": f"{context}"
                            }
                        ]
                    },
                    {
                        "role": "user",
                        "content": [
                            {
                                "type": "text",
                                "text": ""
                            }
                        ]
                    },
                    {
                        "role": "assistant",
                        "content": [
                            {
                                "type": "text",
                                "text": ""
                            }
                        ]
                    },
                    {
                        "role": "user",
                        "content": [
                            {
                                "type": "text",
                                "text": f"{prompt} + {image_filename}"
                            },
                            {
                                "type": "image_url",
                                "image_url": {
                                    "url": f"data:image/jpeg;base64,{encoded_image}"
                                }
                            },
                            {
                                "type": "text",
                                "text": "\n"
                            }
                        ]
                    }
                ],
                "temperature": 0,
                "top_p": 1,
                "max_tokens": 100
            }

            try:
                response = requests.post(ENDPOINT, headers=headers, json=payload)
                response.raise_for_status()
                response_json = response.json()
                content = response_json.get('choices', [])[0].get('message', {}).get('content', '')
                print(f"Response for {image_filename}: {content}")

                # Append the result to the DataFrame
                df = pd.concat([df, pd.DataFrame({"filename": [image_filename], "response": [content]})], ignore_index=True)
                
            except requests.RequestException as e:
                print(f"Failed to make the request for {image_filename}. Error: {e}")
                # Append "fails" to the DataFrame
                df = pd.concat([df, pd.DataFrame({"filename": [image_filename], "response": ["fails"]})], ignore_index=True)
                break

            time.sleep(20)
    # Save the DataFrame to a CSV file
    df.to_csv(output_csv, index=False)

# Example usage
folder_path = r"C:\Users\hagraham\OneDrive\Northwestern\458GeneralAI\Project\Code\subsetOFImages\resizedReRuntwo"
# prompt = "Describe the image"
output_csv = r"C:\Users\hagraham\OneDrive\Northwestern\458GeneralAI\Project\Results\secondPrompExtraTwo.csv"

query_images_in_folder(folder_path, output_csv)
```
